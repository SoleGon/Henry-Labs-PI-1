```python
## Importamos las librerias necesarias para la Normalizacion

import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
## Leo el CSV
amazon= pd.read_csv('amazon_prime_titles-score.csv')
```

```python
## Con un Head puedo ver la data
amazon.head()
```

```python
## Primer transformacion: agrego la columa id que incluye la primer letra de la plataforma mas showid
amazon['id']= 'a'+amazon['show_id']
```

```python
## Verifico la columna ingresada
amazon.head()
```

```python
## Listo los valores nulos para chequearlos antes de quitarlos
amazon.isna().sum ()
```

```python
## Reemplazo los valores nulos en las columna rating por 'G'
amazon['rating']= amazon ['rating'].fillna('G')
```

```python
## Visualizo el cambio 
amazon.head()
```

```python
## Ahora chequeo que no queden NAN en la columna rating
amazon.isna().sum ()
```

```python
## Miro el tipo de dato de la columna date_added que es la que contiene fechas y veo que es objet
amazon.info()
```

```python
## Paso la columna date_added a fecha
amazon['date_added']= pd.to_datetime(amazon['date_added'])
```

```python
## Visualizo el cambio
amazon.head()
```

```python
## Lo chequeo nuevamente para ver que cambie el tipo de dato
amazon.info()
```

```python
## Ahora paso los campos del texto a minusculas
## Todo lo que es de tipo objet paso a minuscula

amazon['show_id']= amazon['show_id'].str.lower()
amazon['type']= amazon['type'].str.lower()
amazon['title']= amazon['title'].str.lower()
amazon['director']= amazon['director'].str.lower()
amazon['cast']= amazon['cast'].str.lower()
amazon['country']= amazon['country'].str.lower()
amazon['rating']= amazon['rating'].str.lower()
amazon['duration']= amazon['duration'].str.lower()
amazon['listed_in']= amazon['listed_in'].str.lower()
amazon['description']=amazon['description'].str.lower()
amazon['id']= amazon['id'].str.lower()

```

```python
## Visualizo el Cambio
amazon.head()
```

```python
## El campo duracion debe convertirse en dos campos: duration_inty duration_type. El primero será un entero y 
# el segundo una cadena indicando la unidad de medición de duración: 
# min (minutos) o temporada (temporadas)

amazon[['duration_int', 'duration_type']] = amazon['duration'].str.split(' ', n=-1, expand=True)
amazon= amazon[amazon.columns[:10].tolist() + ['duration_int', 'duration_type'] + amazon.columns[10:-2].tolist()]
```

```python
## Visualizo la transformacion anterior
amazon[['duration_int', 'duration_type']].head()
```

```python
## Cambio el tipo de dato de duration_int a entero

amazon['duration_int']=amazon['duration_int'].astype ('int64')
```

```python
## Visualizo el tipo de dato

amazon.info()
```

```python
## Visualizo la data
amazon.head()
```

Aca chequeo los valores existentes en la columna duration_type

```python
## Aca chequeo los valores existentes en la columna duration_type
amazon['duration_type'].unique()
```

```python
## Aca reemplazo seasons y le saco la "s"
amazon['duration_type']= amazon['duration_type'].replace ({'seasons':'season'})
```

```python
## verifico que ahora solo existan dos
amazon['duration_type'].unique()
```

```python
## Visualizo la data
amazon.head()
```

```python
## verifico el tipo de datos
amazon.info()
```

```python
## Aplico todas las transformaciones para la plataforma disney


disney= pd.read_csv('disney_plus_titles-score.csv')
disney['id']= 'd'+disney['show_id']
disney['rating']= disney ['rating'].fillna('G')
disney['date_added']= pd.to_datetime(disney['date_added'])
disney['show_id']= disney['show_id'].str.lower()
disney['type']= disney['type'].str.lower()
disney['title']= disney['title'].str.lower()
disney['director']= disney['director'].str.lower()
disney['cast']= disney['cast'].str.lower()
disney['country']= disney['country'].str.lower()
disney['rating']= disney['rating'].str.lower()
disney['duration']= disney['duration'].str.lower()
disney['listed_in']= disney['listed_in'].str.lower()
disney['description']=disney['description'].str.lower()
disney['id']= disney['id'].str.lower()
disney[['duration_int', 'duration_type']] = disney['duration'].str.split(' ', n=-1, expand=True)
disney= disney[disney.columns[:10].tolist() + ['duration_int', 'duration_type'] + disney.columns[10:-2].tolist()]
disney['duration_int']=disney['duration_int'].astype ('int64')
disney[['duration_int', 'duration_type']].head()
disney['duration_type']=disney['duration_type'].replace ({'seasons':'season'})

```

```python
## Verifico el tipo de dato
disney.info()
```

```python
## Realizo las transformaciones para la plataforma Hulu

hulu= pd.read_csv('hulu_titles-score (2).csv')

hulu['id']= 'h'+hulu['show_id']
hulu['rating']= hulu['rating'].fillna('G')
hulu['date_added']= pd.to_datetime(hulu['date_added'])
hulu['cast']=hulu['cast'].astype (object)


## Es necesario cambiar el tipo de dato de 'cast' para poder pasar todo a minusculas
```

```python
hulu['show_id']= hulu['show_id'].str.lower()
hulu['type']=hulu['type'].str.lower()
hulu['title']= hulu['title'].str.lower()
hulu['director']= hulu['director'].str.lower()
hulu['cast']= hulu['cast'].str.lower()
hulu['country']= hulu['country'].str.lower()
hulu['rating']= hulu['rating'].str.lower()
hulu['duration']= hulu['duration'].str.lower()
hulu['listed_in']= hulu['listed_in'].str.lower()
hulu['description']=hulu['description'].str.lower()
hulu['id']= hulu['id'].str.lower()
```

```python
hulu[['duration_int', 'duration_type']] = hulu['duration'].str.split(' ', n=-1, expand=True)
hulu= hulu[hulu.columns[:10].tolist() + ['duration_int', 'duration_type'] +hulu.columns[10:-2].tolist()]
hulu[['duration_int', 'duration_type']].head()
hulu['duration_int']=hulu['duration_int'].fillna(0)
hulu['duration_int']=hulu['duration_int'].astype ('int64')
hulu['duration_type']=hulu['duration_type'].replace ({'seasons':'season'})
```

```python
## Chequeo el tipo de dato

hulu.info()
```

```python
## Ahora realizo las transformaciones para la plataforma Netflix

netflix= pd.read_csv('netflix_titles-score.csv')
netflix['id']= 'n'+netflix['show_id']
netflix['rating']= netflix['rating'].fillna('G')
netflix['date_added']= pd.to_datetime(netflix['date_added'])
netflix['show_id']= netflix['show_id'].str.lower()
netflix['type']= netflix['type'].str.lower()
netflix['title']= netflix['title'].str.lower()
netflix['director']= netflix['director'].str.lower()
netflix['cast']= netflix['cast'].str.lower()
netflix['country']= netflix['country'].str.lower()
netflix['rating']= netflix['rating'].str.lower()
netflix['duration']= netflix['duration'].str.lower()
netflix['listed_in']= netflix['listed_in'].str.lower()
netflix['description']=netflix['description'].str.lower()
netflix['id']= netflix['id'].str.lower()
netflix[['duration_int', 'duration_type']] = netflix['duration'].str.split(' ', n=-1, expand=True)
netflix= netflix[netflix.columns[:10].tolist() + ['duration_int', 'duration_type'] + netflix.columns[10:-2].tolist()]
netflix[['duration_int', 'duration_type']].head()
netflix['duration_int']=netflix['duration_int'].fillna(1)
netflix['duration_int']=netflix['duration_int'].astype ('int64')
netflix['duration_type']=netflix['duration_type'].replace ({'seasons':'season'})
```

```python
## Verifico el tipo de datos

netflix.info()
```

```python
## Concateno los 3 dataframes en uno solo

df= pd.concat([amazon,disney, hulu, netflix ], axis= 0)
```
